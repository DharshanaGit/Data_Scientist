```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
#Training data(Hours studied,Marks Scored)
X=np.array([[2],[4],[6],[1]]) #Hours studied
y=np.array(['Pass','Pass','Pass','Fail']) #result
```

```python
print(X) #2D
print(y) #Vector
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
new_data=[[3]]
prediction=model.predict(new_data)
print("Predicted result for 3 Hours study: ",prediction[0])
```

```python
plt.figure(figsize=(8,5))
for i in range(len(X)):
    if y[i]=='Pass':
        plt.scatter(X[i],0,color='green',label='Pass' if i == 0 else "")
    else:
        plt.scatter(X[i],0,color='red',label='Fail' if i ==3  else "")
plt.scatter(new_data,0,color='blue',marker='x',s=100,label="New Data(3 Hours)")
plt.xlabel('Hours Studied')
plt.yticks([])
plt.title('KNN Pass or Fail Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
